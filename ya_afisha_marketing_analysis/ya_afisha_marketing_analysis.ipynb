#### Retention Rate


Коэффициент удержания. Посмотрю сколько пользователей возвращаются после первого посещения.  
Чтобы оценить этот показателей, мне надо будет выделить временную когорту (у меня это будет месяц, так как данные представлены за целый год), для которого я расчитаю количество новых пользователей, а потом отслежу их отток

```python
df_visits
```

```python
df_first_visit = df_visits.groupby('uid').agg({'start_ts': 'min'}).reset_index()
```

```python
df_first_visit.columns = ['uid', 'first_visit']
df_first_visit['first_activity_month'] = df_first_visit['first_visit'].astype('datetime64[M]')
```

```python
df_visits = df_visits.merge(df_first_visit, on='uid')
```

Получили столбец с данными о первом посещении для каждого пользователя.

```python
df_visits['activity_month'] = df_visits['start_ts'].astype('datetime64[M]')
```

```python
df_visits['cohort_lifetime_month'] = (df_visits['activity_month'] - df_visits['first_activity_month']) / np.timedelta64(1, 'M')
df_visits['cohort_lifetime_month'] = df_visits['cohort_lifetime_month'].round().astype('int')
```

Выделил столбец с продолжительностью жизни для когорты в количестве месяцев

```python
cohorts = df_visits.groupby(['first_activity_month','cohort_lifetime_month']).agg({'uid':'nunique'}).reset_index()
```

Сгруппировав по первому месяцу посещения я выделил когорту и сгруппирововав по продолжительности жизни этой когорты я расчитываю количество уникальных пользователей, таким образом можно отследить отток числа пользователей каждой когорты

```python
initial_users_count = cohorts[cohorts['cohort_lifetime_month'] == 0][['first_activity_month', 'uid']]
```

```python
initial_users_count.columns = ['first_activity_month', 'cohort_users']
```

```python
cohorts = cohorts.merge(initial_users_count, on='first_activity_month')
```

Здесь я определяю изанчальное количество пользователей для каждой когорты и добавляю этот столбец к нашей таблице. Посмотрим что получилось

```python
cohorts
```

Недостает только нашего искомого показателя, добавим его и построим таблицу и график

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
```

```python
retention_pivot = cohorts.pivot_table(
    index='first_activity_month',
    columns='cohort_lifetime_month',
    values='retention'
) 
```

```python
retention_pivot
```

Избавимся от 0 столбца, так как мы считаем относительный показатель и никакой информации он не несет, а только будет искажать картину на heatmap'е.

```python
retention_pivot = retention_pivot.drop(0, axis='columns')
```

```python
sns.set(style='white')
plt.figure(figsize=(25, 20))
plt.title('Retention пользователей по месячным когортам')
sns.heatmap(retention_pivot, annot=True, fmt='.1%');
```

```python
retention_rate_for_2_months = retention_pivot.loc[:, 2].mean()
retention_rate_for_2_months
```

```python
print('Средний Retention Rate на второй месяц «жизни» когорт = {:.2%}'.format(retention_rate_for_2_months))
```

Основной отток пользователей происходит после первого месяца и уже ко второму месяцу остается только в среднем 6,52% от первоначально привлеченных пользователей. Потом же отток не так существен. Можно было бы поработать над удержанием клинетов, хоть и специфика сайта предполагает нерегулярность посещений, а только по мере необходимости, которая может возникать не часто (покупка билетов в театры, на концерты, квесты и т.п.). С другой стороны в кино люди ходят часто и можно было бы работать на удержание клиентов в этом направлении. #### Среднее количество покупок на одного покупателя за 6 месяцев

Проделаем те же операции для выделения когорт, на этот раз с данными о покупках.

```python
df_orders['order_month'] = df_orders['buy_ts'].astype('datetime64[M]')
```

```python
first_orders = df_orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
```

```python
first_orders.columns = ['uid', 'first_order_month']
```

```python
df_orders = df_orders.merge(first_orders, on='uid')
```

```python
order_cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
```

```python
order_cohort_sizes.columns = ['first_order_month', 'n_buyers']
```

```python
order_cohorts = (
    df_orders.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'count'})
    .reset_index()
)
```

```python
order_cohorts.columns = ['first_order_month', 'order_month', 'orders_cnt']
```

```python
report = order_cohorts.merge(order_cohort_sizes, on='first_order_month')
```

```python
report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
```

```python
report['age'] = report['age'].round().astype('int')
```

```python
report['order_per_buyer'] = report['orders_cnt'] / report['n_buyers']
```

```python
output = report.pivot_table(
    index='first_order_month', 
    columns='age', 
    values='order_per_buyer', 
    aggfunc='mean'
)

output.cumsum(axis=1).round(2).fillna('')  
```

```python
output.cumsum(axis=1)[6].mean()
```

В первый месяц в среднем покупатели совершали чуть больше одного заказа, то есть окол 15% совершали два заказа за месяц (грубоговоря), потом пользователей становилось меньще и соответственно меньше покупок, но так как они накапливаются и совершаются изначально привлеченными покупателями, то у нас получается, что к 6 месяцу в среднем на одного покупателя совершается 1,5 покупки, или 3 покупки на 2 покупателей, или 15 покупок на 100 покупателей, или... В общем не много, после первой покупки, если честно.

#### Изменение LTV на покупателя по когортам

Вновь когорты, но уже другие показатели. На этот раз LTV, то есть итоговая сумма, которую мы получаем в среднем с одного клиента за весь период его пользования сервисом.

```python
ltv_cohorts = (
    df_orders.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)
```

```python
report2 = ltv_cohorts.merge(order_cohort_sizes, on='first_order_month')
```

```python
margin_rate = 1
```

Хотя у нас маржинальность 100% и валовая прибыль (gross profit) получается равна выручке (revenue), мы для пластичности кода произведем расчеты по всем ступеням

```python
report2['gp'] = report2['revenue'] * margin_rate
```

```python
report2['age'] = (
    report2['order_month'] - report2['first_order_month']
) / np.timedelta64(1, 'M')
report2['age'] = report2['age'].round().astype('int')
```

```python
report2['ltv'] = report2['gp'] / report2['n_buyers']
```

```python
output2 = report2.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
).cumsum(axis=1)

output2.fillna('') 
```

```python
output2.loc[:, 6].mean()
```

```python
sns.set(style='white')
plt.figure(figsize=(17, 8))
plt.title('LTV')
sns.heatmap(output2, annot=True, fmt='.2f');
```

Heatmap наглядно показывает нам, что первая (июньская) и третья (сентябрьская) когорты приносят больше денег от покупателей (возможно они были более правильно привлечены или простимулированы). На 6 месяц когорты средний ltv на одного покупателя составляет 8,4 у.е. Хорошо это или плохо? Надо понять по расходам на привлечение. Это далее

#### ROMI по когортам в разрезе источников

Попробую все с самого начала еще раз прогнать

```python
df_orders
```

```python
df_visits.groupby('uid')['first_source_id'].mean().reset_index()
```

Добавили для каждого польщователя информацию об источники привлчения. Теперь присоединим эту информацию к данным о заказах

```python
df_orders_new = df_orders.merge(df_visits.groupby('uid')['first_source_id'].mean().reset_index(), how='left', on='uid')
df_orders_new
```

Выделяем когорты по месяцу и ресурсу и считаем количество покупателей

```python
cohort_sizes2 = df_orders_new.groupby(['first_order_month', 'first_source_id']).agg({'uid': 'nunique'}).reset_index()
```

```python
cohort_sizes2.columns = ['first_order_month', 'first_source_id', 'cnt_buyers']
```

```python
cohorts2 = df_orders_new.groupby(['first_order_month', 'first_source_id', 'order_month']).agg({'revenue': 'sum'}).reset_index()
```

```python
cohorts2 = cohorts2.merge(cohort_sizes2, on=['first_order_month', 'first_source_id'])
```

Добавляем показатели по валовой прибыли

```python
margin_rate = 1
```

```python
cohorts2['gp'] = cohorts2['revenue'] * margin_rate
```

И возрасту когорты

```python
cohorts2['age'] = (cohorts2['order_month'] - cohorts2['first_order_month']) / np.timedelta64(1, 'M')
```

```python
cohorts2['age'] = cohorts2['age'].round().astype('int')
```

Расчитаем общий принесенный доход с клиента

```python
cohorts2['ltv'] = cohorts2['gp'] / cohorts2['cnt_buyers']
```

```python
cohorts2
```

Теперь не хватает только затрат на привлечение

```python
df_costs_by_month = df_costs.groupby(['month', 'source_id']).agg({'costs': 'sum'}).reset_index()
df_costs_by_month
```

```python
report3 = cohorts2.merge(df_costs_by_month, left_on=['first_order_month', 'first_source_id'], right_on=['month', 'source_id'])
```

```python
report3['cac'] = report3['costs'] / report3['cnt_buyers']
```

И наконец ROMI (Return on Marketing Investment) - какой возврат мы получаем от наших вложений в привлечение клиентов.

```python
report3['romi'] = report3['ltv'] / report3['cac']
```

Построим графики для каждого источника в отдельности

```python
report3['source_id'].value_counts()
```

```python
for source in [1,2,3,4,5,9,10]:
    output3 = report3[report3['source_id'] == source].pivot_table(
                index='first_order_month',
                columns='age',
                values='romi',
                aggfunc='mean').cumsum(axis=1)
    sns.set(style='white')
    plt.figure(figsize=(17, 8))
    plt.title('ROMI для ресурса № {}'.format(source))
    sns.heatmap(output3, annot=True, fmt='.2f');
    
```

Вот теперь самое инетерсное. Во-первых наш излюбленный 3 ресурс имеет ужасные показатели и очень низкую отдачу, с этим определенно надо что-то делать. Во-вторых у нас очень высокие показатели по первому ресурсу, в который определенно стоит вкладываться. По 10-му можно даже не смотреть, непопулярный и с низкими показателями. Явно выбиваются сентябрьская когорта для источника 2 и августовская по источнику 9. Нужно понять что за рекламная кампания проводилась в это время на этих ресурсах и попробовать повторить этот успех :) 5 и 4 источники более равномерны и выходят в плюс, работаю как-будто в паре, хотя стоит обратить внимание на них, потому что показатели даже по первым месяцам под конец отчетного периода стали ухудшаться...
